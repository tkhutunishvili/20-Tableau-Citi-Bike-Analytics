```python
# Combine all data
citibike_201806_12 = pd.concat([JC_201806_citibike_tripdata_df, JC_201807_citibike_tripdata_df, JC_201808_citibike_tripdata_df, JC_201809_citibike_tripdata_df, JC_201810_citibike_tripdata_df, JC_201811_citibike_tripdata_df, JC_201812_citibike_tripdata_df])
citibike_201806_12.tail()
```

```python
#Gender column cleanup/rename
citibike_201806_12 = citibike_201806_12.replace({0: None})
citibike_201806_12 = citibike_201806_12[citibike_201806_12.gender.notna()]
citibike_201806_12['gender'].head(20)
citibike_201806_12['gender'] = citibike_201806_12['gender'].replace({1: 'male', 2: 'female'})
citibike_201806_12['gender'].head()
```
